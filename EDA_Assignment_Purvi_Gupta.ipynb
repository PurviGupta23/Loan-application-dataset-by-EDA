# Checking Data Imbalance

```python
# Listing columns for checking data imbalance and plotting them
col_list = ['TARGET','NAME_CONTRACT_TYPE','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','NAME_INCOME_TYPE',
            'NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE']
k=0
plt.figure(figsize=(20,15))
for col in col_list:    
    k=k+1
    plt.subplot(3, 3,k)    
    app_data[col].value_counts().plot(kind='bar');
    plt.title(col)
```

We can see that there is data imbalance in below columns:-

- TARGET - There are very few defaulters(1) compare to non defaulters(0)
- NAME_CONTRACT_TYPE - There are very few Revolving loans than Cash loans
- NAME_EDUCATION_TYPE - Most of the loans applied by Secondary/Secondary special educated people
- NAME_FAMILY_STATUS - Most of the loans applied by Married people.
- NAME_HOUSING_TYPE - Most of the application came from Home/appartment owner

```python
# creating three age group
def age_group(x):
    if (x < 40):
        return 'Young'
    elif (x >= 40 and x < 60):
        return 'Mid Age'
    else:
        return 'Senior'
```

```python
# creating new column for age
app_data['age_category'] = app_data['age'].apply(age_group)
app_data.head()
```

```python
# creating three credit amount group
def credit_group(x):
    if (x < 500000):
        return 'Low'
    elif (x >= 500000 and x < 750000):
        return 'Medium'
    else:
        return 'High'
```

```python
app_data['credit'] = app_data['AMT_CREDIT'].apply(credit_group)
app_data.head()
```

```python
# creating three income group
def income_group(x):
    if (x < 100000):
        return 'Low'
    elif(x >= 100000 and x < 150000):
        return 'Medium'
    else:
        return 'High'
```

```python
app_data['income'] = app_data['AMT_INCOME_TOTAL'].apply(income_group)
app_data.head()
```

```python
# Finding outliers in all the numerical columns with 1.5 IQR rule and removing the outlier records 
col_list_outlier = ['AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_GOODS_PRICE','DAYS_DECISION','CNT_PAYMENT']
for col in col_list_outlier: 
    q1 = pre_data[col].quantile(0.25)
    q3 = pre_data[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    pre_data_p = pre_data.loc[(pre_data[col] > range_low) & (pre_data[col] < range_high)]

pre_data_p.shape
```

- NAME_PORTFOLIO - Very few application for Cards and Cars
- CHANNEL_TYPE - Except Country-Wide, Credit and Cash offices and Stone all other channels are very few in number.
